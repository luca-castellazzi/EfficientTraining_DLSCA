```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ges_1d = np.load('/prj/side_channel/Pinata/results/DKTA/SBOX_OUT/1d/avg_ge__ga.npy')
ges_2d = np.load('/prj/side_channel/Pinata/results/DKTA/SBOX_OUT/2d/avg_ge__ga.npy')

ges_tot = np.array([ges_1d, ges_2d])
```

```python
f, ax = plt.subplots(1, 2, figsize=(20,5), sharex=True, sharey=True)

for i in range(ges_tot.shape[0]):
    for ge in ges_tot[i]:
        ax[i].plot(ge)
    
plt.show()
```

```python
x = range(2)
y = range(10)
X, Y = np.meshgrid(y, x)

fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection ='3d')

for n_traces in range(10):
    z = np.array([ges_tot[i][j][n_traces] 
                  for i in x 
                  for j in y]).reshape((2,10))
    ax.plot_surface(X, Y, z)
    
    

# Z = ges_tot[tuple(np.meshgrid(x, y))]


# ax.plot_wireframe(Y, X, z)
# ax.plot_surface(X, Y, z, cmap='plasma')
# ax.contour3D(X, Y, z, cmap='plasma')
# ax.scatter(X, Y, z)

plt.show()
```
