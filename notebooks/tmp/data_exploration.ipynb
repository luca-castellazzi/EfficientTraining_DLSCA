```python
# Basics
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import random
import seaborn as sns

# Custom
import sys
sys.path.insert(0, '../src/utils')
from data_loader import DataLoader
import constants
```

```python
BYTE_IDX = 0

all_labels = []
for d in constants.DEVICES:
    for k in constants.KEYS.keys():
        dl = DataLoader(f'/prj/side_channel/PinataTraces/datasets/SBOX_OUT/{d}-{k}.json', BYTE_IDX, train_perc=1.0)
        _, ohe_labels = dl.gen_train()
        labels = [l.tolist().index(1) for l in tqdm(ohe_labels)]
        
        all_labels.append(labels)
```

```python
plt.hist(all_labels[0])
```

```python
f, ax = plt.subplots(figsize=(15, 10))
sns.histplot(all_labels[:], ax=ax, kde=True, discrete=True)
```
